```python
temp = pd.read_csv("./data/patient_symptoms.csv", delimiter=";")
temp = temp["Date"].str.split(",", expand=True)
df["Date"] = temp[0]
df["Time"] = temp[1]
df["Date"] = pd.to_datetime(df["Date"])
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')
df.head()
```
